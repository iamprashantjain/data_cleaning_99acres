```python
#columns to drop
df.drop(columns=['link','property_id'],axis=1,inplace=True)
```

```python
#data cleaning starts from here
df.sample(3)
```

```python
#rename column name
df.rename(columns={'area':'price_per_sqft'}, inplace=True)
```

```python
#society column - always run value_counts() on columns to take a quick glance while cleaning
df['society'].value_counts()
```

```python
#remove ratings info as rating column is already in data & 
#convert to lower() to avoid duplicate socities
#avoiding ratings info bcoz ratings not availble for lot of socities
import re
df['society'] = df['society'].apply(lambda name: re.sub(r'\d+(\.\d+)?\s?★','',str(name)).strip()).str.lower()
```

```python
df['society'].value_counts()
```

```python
df['society'].value_counts().shape
```

```python
#cleaning price column
#there are 2 types of prices.. in crore & in lakhs
#need to standardize the price columns in same scale
#also there are 11 socities where price says: "Price on Request"
#so we'll remove them bcoz it doesnt make any sense for price prediction model

df['price'].value_counts()
```

```python
df = df[df['price'] != "Price on Request"]
```

```python
df.sample(5)
```

```python
#if prices in lac then converting them to float & crores
#else converting them into float

def convert_price(x):
    if type(x) == float:
        return x
    else:
        if x[1] == 'Lac':
            return round(float(x[0])/100,2)
        else:
            return round(float(x[0]),2)
```

```python
df['price'] = df['price'].str.split(' ').apply(convert_price)
```

```python
df.sample(5)
```

```python
#cleaning price_per_sqft

df['price_per_sqft'].value_counts()
```

```python
df['price_per_sqft'] = df['price_per_sqft'].str.replace("/sq.ft.", "").str.replace("₹","").str.replace(",","").astype('float')
```
